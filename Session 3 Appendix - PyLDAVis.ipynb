# Session 3 Appendix: PyLDAVis

This appendix demonstrates the PyLDAVis utility for visualizing LDA topic models. 

```python
from gensim.corpora.dictionary import Dictionary 
from gensim.models.ldamodel import LdaModel
from gensim.corpora import MmCorpus

id2word = Dictionary.load("data/corpus.dict")
mmcorpus = MmCorpus("data/corpus.mm")
lda = LdaModel.load("data/corpus.lda")
```

```python
import pyLDAvis
import pyLDAvis.gensim

pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(lda, mmcorpus, id2word)
```
